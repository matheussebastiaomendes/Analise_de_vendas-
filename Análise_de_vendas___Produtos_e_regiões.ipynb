# Cidades por produto por número de vendas

```python
contagem_vendas_cidade_produto = df.groupby(['Product', 'Cidade'])['Order ID'].count().reset_index()
contagem_vendas_cidade_produto.rename(columns={'Order ID': 'Quantidade Vendas'}, inplace=True)
contagem_vendas_cidade_produto
```

```python
contagem_vendas_cidade_produto = contagem_vendas_cidade_produto.sort_values(by=['Product', 'Quantidade Vendas'], ascending=[True, False])
contagem_vendas_cidade_produto
```

## 3 Melhores cidades por cada produto


```python
top_cidades_por_produto = pd.DataFrame()
produtos_unicos = contagem_vendas_cidade_produto['Product'].unique()
for produto in produtos_unicos:
    produto_df = contagem_vendas_cidade_produto[contagem_vendas_cidade_produto['Product'] == produto]
    top_cidades = produto_df.head(3)[['Cidade', 'Quantidade Vendas']]
    top_cidades['Product'] = produto
    temp_df = top_cidades[['Product', 'Cidade', 'Quantidade Vendas']]
    top_cidades_por_produto = pd.concat([top_cidades_por_produto, temp_df])
top_cidades_por_produto.reset_index(drop=True, inplace=True)
top_cidades_por_produto
```

## 3 Piores cidades por cada produto


```python
piores_cidades_por_produto = pd.DataFrame()
produtos_unicos = contagem_vendas_cidade_produto['Product'].unique()
for produto in produtos_unicos:
    produto_df = contagem_vendas_cidade_produto[contagem_vendas_cidade_produto['Product'] == produto]
    piores_cidades = produto_df.tail(3)[['Cidade', 'Quantidade Vendas']]
    piores_cidades['Product'] = produto
    temp_df = piores_cidades[['Product', 'Cidade', 'Quantidade Vendas']]
    piores_cidades_por_produto = pd.concat([piores_cidades_por_produto, temp_df])
piores_cidades_por_produto.reset_index(drop=True, inplace=True)
piores_cidades_por_produto
```

# Cidades por Produto por Lucro

```python
cidade_produto_lucro = df.groupby(['Product', 'Cidade'])['margin'].sum().reset_index()
cidade_produto_lucro.rename(columns={'margin': 'Lucro'}, inplace=True)
cidade_produto_lucro
```

```python
ordem_cidade_por_lucro = cidade_produto_lucro.sort_values(by=['Product', 'Lucro'], ascending=[True, False])
ordem_cidade_por_lucro
```

## 3 Melhores Cidades por cada Produto por Lucro



```python

top_cidades_por_produto_por_lucro = pd.DataFrame()
produtos_unicos = ordem_cidade_por_lucro['Product'].unique()
for produto in produtos_unicos:
    produto_df = ordem_cidade_por_lucro[ordem_cidade_por_lucro['Product'] == produto]
    top_cidades = produto_df.head(3)[['Cidade', 'Lucro']]
    top_cidades['Product'] = produto
    temp_df = top_cidades[['Product', 'Cidade', 'Lucro']]
    top_cidades_por_produto_por_lucro = pd.concat([top_cidades_por_produto_por_lucro, temp_df])
top_cidades_por_produto_por_lucro.reset_index(drop=True, inplace=True)
top_cidades_por_produto_por_lucro
```

## 3 Piores Cidades por cada Produto por Lucro

```python
piores_cidades_por_produto_por_lucro = pd.DataFrame()
produtos_unicos = ordem_cidade_por_lucro['Product'].unique()
for produto in produtos_unicos:
    produto_df = ordem_cidade_por_lucro[ordem_cidade_por_lucro['Product'] == produto]
    piores_cidades = produto_df.tail(3)[['Cidade', 'Lucro']]
    piores_cidades['Product'] = produto
    temp_df = piores_cidades[['Product', 'Cidade', 'Lucro']]
    piores_cidades_por_produto_por_lucro = pd.concat([piores_cidades_por_produto_por_lucro, temp_df])
piores_cidades_por_produto_por_lucro.reset_index(drop=True, inplace=True)
piores_cidades_por_produto_por_lucro
```
